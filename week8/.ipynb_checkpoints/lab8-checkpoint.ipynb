## CountVectorizing the training text using klearn.feature_extraction.text

```python
from sklearn.feature_extraction.text import CountVectorizer
clf = CountVectorizer(max_features=1200)
trained_text = clf.fit_transform(df_train['text'])
trained_text = trained_text.toarray()
trained_text
```

## print the unique features of training text

```python
clf.get_feature_names_out()
```

## distinguish X-train and y-train

```python
X = trained_text
y = df_train['target']
```

## create the Naive Bayes classifier

```python
from sklearn.naive_bayes import BernoulliNB
NBclf = BernoulliNB()
NBclf.fit(X,y)
```

## read the test data set and display

```python
#now prediction we need test so ...
df_test = pd.read_csv('newsgroups_test.csv')
df_test = df_test
df_test
```

## set the y-test

```python
y = df_test['target']
y
```

## CountVectorizing the testing text using klearn.feature_extraction.text

```python
test_text = clf.fit_transform(df_test['text'])
test_text = test_text.toarray()
test_text
```

## print the unique features of training text

```python
clf.get_feature_names_out()
```

## shape of testing text data

```python
test_text.shape
```

## metrics evaluation 

```python
from sklearn.metrics import accuracy_score,classification_report
pred = NBclf.predict(test_text)

accuracy = accuracy_score(y,pred)
report = classification_report(y,pred)
print("\nAccuracy is : ",accuracy ,"\n\nThe report is : \n ",report)
      
```

## remove the stop words and evaluating

```python
from sklearn.feature_extraction.text import CountVectorizer
clf_stop = CountVectorizer(max_features=1500,stop_words='english')
trained_final = clf_stop.fit_transform(df_train.text)
trained_final = trained_final.toarray()
trained_final
```

```python
from sklearn.naive_bayes import BernoulliNB
NBclf = BernoulliNB()
NBclf.fit(trained_final,df_train.target)
```

```python
test_text_final = clf_stop.fit_transform(df_test.text)
test_text_final = test_text_final.toarray()
test_text_final
```

```python
from sklearn.metrics import accuracy_score,classification_report
pred_final = NBclf.predict(test_text_final)

accuracy_final = accuracy_score(y,pred_final)
report_final = classification_report(y,pred_final)
print(f"\nAccuracy is : {accuracy} \n\nThe report is : \n {report}")
```
