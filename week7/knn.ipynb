```python
# 10
def show_img(x,y,num=5):
    for i in range(num):
        plt.subplot(1,num,i+1)
        plt.imshow(x[i].reshape(28,28))
        plt.axis('off')
    plt.show()
    
show_img(x.values,y.values,5)
```

```python
#11 split train=75 validation 25%
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25, random_state=42,stratify=y)
```

```python

from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=3,metric='euclidean')
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report,ConfusionMatrixDisplay,accuracy_score
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
#Decison tree
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
# visualize the image of first 5 digits in x-test and also print the predicted
import matplotlib.pyplot as plt
import numpy as np

# Predict the labels for the first 5 test samples
predictions = model.predict(x_test[:5])

# Plot the first 5 test images with their predicted labels
plt.figure(figsize=(10, 2))
for i in range(5):
    image = np.array(x_test.iloc[i]).reshape(28, 28)  # reshape to 28x28 pixels
    plt.subplot(1, 5, i + 1)
    plt.imshow(image, cmap='gray')
    plt.title(f'Pred: {predictions[i]}')
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
#Apply prediction on any one sample input and print the probability estimates for each class 
# Pick a single test sample (e.g., the first one)
sample = x_test.iloc[0].values.reshape(1, -1)

# Predict probabilities for all classes (0â€“9)
probs = model.predict_proba(sample)

# Print the predicted probabilities
for i, prob in enumerate(probs[0]):
    print(f"Class {i}: {prob:.4f}")
```

```python
predicted_label = model.predict(sample)[0]
actual_label = y_test.iloc[0]

print(f"\nPredicted Label: {predicted_label}")
print(f"Actual Label: {actual_label}")
```
