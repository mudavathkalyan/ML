# **<font color='darkorange'>Regression Tutorial</font>**

Regression is a fundamental technique in **Machine Learning** used to model relationships between variables. It is widely applied to predict outcomes based on input data.

Regression can be broadly categorized into:  
1. **<font color='blue'>Linear Regression</font>**: Assumes a linear relationship between variables.  
2. **<font color='green'>Polynomial Regression</font>**: Captures non-linear relationships by introducing polynomial terms.

---

## **<font color='blue'>1. Linear Regression</font>**

Linear regression assumes a **straight-line relationship** between the independent variable(s) and the target variable.

---

### **<font color='blue'>Linear Regression Model</font>**

The general equation for a linear regression model is:  

$$
\hat{y} = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \cdots + \theta_n x_n
$$

Where:  
- **$ \hat{y} $**: The predicted value.  
- **$ n $**: The number of features.  
- **$ x_i $**: The $i$-th feature value.  
- **$ \theta_j $**: The $j$-th model parameter (including the bias term $ \theta_0 $ and the feature weights $ \theta_1, \theta_2, \dots, \theta_n $).

---

### **<font color='purple'>Simple Linear Regression</font>**  

When there is only one feature ($ n = 1 $), the equation simplifies to:  

$$
\hat{y} = \theta_0 + \theta_1 x_1
$$

- **<font color='darkcyan'>Practical Example</font>**:  
  A company uses **experience** as the independent variable to predict **salary**.  
  - **$ x_1 $**: Experience  
  - **$ \hat{y} $**: Predicted Salary  
  - **Goal**: Understand how experience impacts salary.

---

### **<font color='purple'>Multiple Linear Regression</font>**  

When there are multiple features ($ n > 1 $), the equation becomes:  

$$
\hat{y} = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \cdots + \theta_n x_n
$$

- **<font color='darkcyan'>Practical Example</font>**:  
  A real estate agency uses **property features** to predict the **house price**.  
  - **$ x_1 $**: Square footage  
  - **$ x_2 $**: Number of bedrooms  
  - **$ x_3 $**: Location rating  
  - **$ \hat{y} $**: Predicted House Price  
  - **Goal**: Analyze how multiple factors affect housing prices.

---

## **<font color='green'>2. Polynomial Regression</font>**

Polynomial regression extends linear regression to model **non-linear relationships** between variables. It transforms the features into polynomial terms.

The equation for Polynomial Regression can be like:  

$$
\hat{y} = \theta_0 + \theta_1 x_1 + \theta_2 x_1^2 + \theta_3 x_1^3 + \cdots + \theta_n x_1^n
$$

Where:  
- **$ \hat{y} $**: Predicted Value.  
- **$ x_1, x_1^2, x_1^3, \dots $**: Polynomial terms.  
- **$ \theta_0, \theta_1, \theta_2, \dots $**: Model parameters.  

- **<font color='darkcyan'>Practical Example</font>**:  
   Predicting the **salary** based on **position level**.  
   - **$ x_1 $**: Position Level  
   - **$ \hat{y} $**: Salary  
   - **Goal**: Use a polynomial regression model to predict the non-linear growth pattern.

## **<font color='darkorange'>Comparison of Linear and Polynomial Regression</font>**

| **<font color='blue'>Aspect</font>**     | **<font color='blue'>Linear Regression</font>**         | **<font color='green'>Polynomial Regression</font>**    |
|------------------------|--------------------------------|------------------------------------|
| **Nature of Relationship** | Assumes a straight-line relationship   | Models non-linear relationships (curved) |
| **Complexity**         | Simple and interpretable                 | More complex due to polynomial terms     |
| **Use Cases**          | When data follows a linear trend         | When data shows curves or non-linear trends |

---

- **<font color='blue'>Linear Regression</font>** is suitable for data with a linear relationship, like predicting sales based on advertising.  
- **<font color='green'>Polynomial Regression</font>** is ideal when the relationship is non-linear, such as growth patterns over time.  

Understanding the type of relationship in the data helps choose the correct regression technique to make accurate predictions. ðŸš€


## **<font color='darkorange'>Methods for Finding Optimal Model Parameters</font>**

When training a machine learning model, there are two primary approaches to find the model parameters that minimize the cost function over the training set:

### **<font color='blue'>1. Closed-Form Solution</font>**
- Directly computes the model parameters that best fit the training set.  
- Finds the parameters that **minimize the cost function** using a mathematical formula.  

---

### **<font color='blue'>2. Iterative Optimization</font>**
An optimization approach that **gradually tweaks the model parameters** to minimize the cost function:

#### **<font color='green'>Gradient Descent (GD)</font>**
- Iteratively adjusts the model parameters in the direction that reduces the cost function.  
- **Goal**: Minimize the cost function over the training set.  

---

#### **Variants of Gradient Descent**  
Gradient Descent has three main variants based on the amount of data used in each step:  

1. **<font color='purple'>Batch Gradient Descent</font>**  
   - Uses the **entire training set** to compute the gradient at each step.  

2. **<font color='purple'>Mini-Batch Gradient Descent</font>**  
   - Splits the training set into **small batches** and computes the gradient for each batch.  

3. **<font color='purple'>Stochastic Gradient Descent (SGD)</font>**  
   - Uses **one training example** at a time to compute the gradient.

# **<font color='darkorange'>Derivation of the Simple Linear Regression Formula</font>**

**Simple Linear Regression (SLR)** aims to model the relationship between one independent variable $x$ and one dependent variable $y$. The relationship is represented as:

$$
\hat{y} = \theta_0 + \theta_1 x
$$

Where:  
- **$\hat{y}$**: Predicted value (dependent variable)  
- **$x$**: Independent variable (predictor)  
- **$\theta_0$**: Intercept (value of $\hat{y}$ when $x = 0$)  
- **$\theta_1$**: Slope (rate of change of $\hat{y}$ with respect to $x$)  

The goal is to determine the values of the slope $\theta_1$ and intercept $\theta_0$ that minimize the **sum of squared residuals (errors)**. Letâ€™s derive the formula step by step.

---

## **<font color='blue'>1. Sum of Squared Residuals</font>**

The **residual (error)** for each data point is the difference between the observed value $y_i$ and the predicted value $\hat{y}_i$:

<img src="https://drive.google.com/uc?id=1-MTdektoO2q1K9JZY-KVGC0TH4mtU267"/>

$$
e_i = y_i - \hat{y}_i
$$

The predicted value $\hat{y}_i$ is given as:

$$
\hat{y}_i = \theta_0 + \theta_1 x_i
$$

Thus, the residual becomes:

$$
e_i = y_i - (\theta_0 + \theta_1 x_i)
$$

The objective is to minimize the **Sum of Squared Errors (SSE)**:

$$
SSE = \sum_{i=1}^n e_i^2 = \sum_{i=1}^n \left(y_i - (\theta_0 + \theta_1 x_i)\right)^2
$$

---

## **<font color='blue'>2. Deriving the Slope ($\theta_1$)</font>**

To minimize **SSE**, we differentiate the **SSE** with respect to $\theta_0$ and $\theta_1$, and set the derivatives to zero.

<img src="https://drive.google.com/uc?id=1puK9mBV2XqNdGM_BGWPvwx_lMxMUYeKz"/>

### **Step 1: Partial Derivative with Respect to $\theta_0$**

Taking the derivative of SSE with respect to $\theta_0$:

$$
\frac{\partial SSE}{\partial \theta_0} = -2 \sum_{i=1}^n \left(y_i - \theta_0 - \theta_1 x_i\right)
$$

Set this derivative to zero:

$$
\sum_{i=1}^n (y_i - \theta_0 - \theta_1 x_i) = 0
$$

Rearranging, we get:

$$
\sum y_i = n\theta_0 + \theta_1 \sum x_i
$$

---

### **Step 2: Partial Derivative with Respect to $\theta_1$**

Taking the derivative of SSE with respect to $\theta_1$:

$$
\frac{\partial SSE}{\partial \theta_1} = -2 \sum_{i=1}^n x_i \left(y_i - \theta_0 - \theta_1 x_i\right)
$$

Set this derivative to zero:

$$
\sum_{i=1}^n x_i (y_i - \theta_0 - \theta_1 x_i) = 0
$$

Expanding and rearranging:

$$
\sum x_i y_i = \theta_0 \sum x_i + \theta_1 \sum x_i^2
$$

---

## **<font color='blue'>3. Solving for $\theta_0$ and $\theta_1$</font>**

We now have two equations:

1. $ \sum y_i = n\theta_0 + \theta_1 \sum x_i $  
2. $ \sum x_i y_i = \theta_0 \sum x_i + \theta_1 \sum x_i^2 $  

### **Solving for $\theta_1$ (Slope)**

First, substitute $\theta_0$ from the first equation into the second equation.

After simplification, the slope $\theta_1$ is given as:

$$
\theta_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}
$$

Where:  
- $\bar{x}$: Mean of $x$ values  
- $\bar{y}$: Mean of $y$ values  

The numerator represents the **covariance** between $x$ and $y$, and the denominator represents the **variance** of $x$.

---

### **Solving for $\theta_0$ (Intercept)**

Once we have $\theta_1$, the intercept $\theta_0$ can be calculated using:

$$
\theta_0 = \bar{y} - \theta_1 \bar{x}
$$

---

## **<font color='blue'>4. Final SLR Formula</font>**

The final Simple Linear Regression equation is:

$$
\hat{y} = \theta_0 + \theta_1 x
$$

Where:  
- **$\theta_1$ (Slope)**:  

$$
\theta_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}
$$  

- **$\theta_0$ (Intercept)**:  

$$
\theta_0 = \bar{y} - \theta_1 \bar{x}
$$

---

## **<font color='darkorange'>Summary</font>**

1. **Slope $\theta_1$** measures the rate of change of $\hat{y}$ with respect to $x$ (calculated as **covariance** of $x$ and $y$ divided by the **variance** of $x$).  
2. **Intercept $\theta_0$** determines the value of $\hat{y}$ when $x = 0$.  
3. Together, $\theta_0$ and $\theta_1$ define the best-fit line that minimizes the sum of squared residuals.

The derived formulas allow us to find the optimal regression line for any given set of data. ðŸš€


## **<font color='darkorange'>Example of Simple Linear Regression</font>**

Let's consider a real-world example of predicting a student's **exam score** based on the number of hours they study.

---

### **<font color='blue'>Problem Statement</font>**

We want to model the relationship between:  
- **<font color='green'>Independent Variable</font>** ($x$): Hours of study  
- **<font color='green'>Dependent Variable</font>** ($y$): Exam score  

The goal is to find the equation of the best-fit line:

$$
\hat{y} = \theta_0 + \theta_1 x
$$

Where:  
- $\hat{y}$: Predicted exam score  
- $x$: Hours of study (independent variable)  
- $\theta_0$: Intercept  
- $\theta_1$: Slope (rate of change of $\hat{y}$ with respect to $x$)  

---

### **<font color='blue'>Data</font>**

| **Hours of Study ($x$)** | **Exam Score ($y$)** |
|--------------------------|----------------------|
| 1                        | 50                   |
| 2                        | 55                   |
| 3                        | 65                   |
| 4                        | 70                   |
| 5                        | 80                   |

---

### **<font color='blue'>Step 1: Calculate the Mean</font>**

1. **Mean of $x$ (Hours of Study):**

$$
\bar{x} = \frac{\sum x_i}{n} = \frac{1 + 2 + 3 + 4 + 5}{5} = 3
$$

2. **Mean of $y$ (Exam Score):**

$$
\bar{y} = \frac{\sum y_i}{n} = \frac{50 + 55 + 65 + 70 + 80}{5} = 64
$$

---

### **<font color='blue'>Step 2: Calculate the Slope ($\theta_1$)</font>**

The formula for slope $\theta_1$ is:

$$
\theta_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}
$$

| **$x_i$** | **$y_i$** | **$x_i$ - $\bar{x}$**  | **$y_i$ - $\bar{y}$** | **($x_i$ - $\bar{x}$)($y_i$ - $\bar{y}$)** | **($x_i$ - $\bar{x}$)$^2$** |
|-----------|-----------|---------------------|---------------------|-------------------------------------|------------------------|
| 1         | 50        | -2                  | -14                 | 28                                  | 4                      |
| 2         | 55        | -1                  | -9                  | 9                                   | 1                      |
| 3         | 65        | 0                   | 1                   | 0                                   | 0                      |
| 4         | 70        | 1                   | 6                   | 6                                   | 1                      |
| 5         | 80        | 2                   | 16                  | 32                                  | 4                      |

1. **Sum of $(x_i - \bar{x})(y_i - \bar{y})$:**

$$
\sum (x_i - \bar{x})(y_i - \bar{y}) = 28 + 9 + 0 + 6 + 32 = 75
$$

2. **Sum of $(x_i - \bar{x})^2$:**

$$
\sum (x_i - \bar{x})^2 = 4 + 1 + 0 + 1 + 4 = 10
$$

Thus, the slope $\theta_1$ is:

$$
\theta_1 = \frac{75}{10} = 7.5
$$

---

### **<font color='blue'>Step 3: Calculate the Intercept ($\theta_0$)</font>**

The formula for the intercept $\theta_0$ is:

$$
\theta_0 = \bar{y} - \theta_1 \bar{x}
$$

Substitute the values:

$$
\theta_0 = 64 - (7.5 \times 3) = 64 - 22.5 = 41.5
$$

---

### **<font color='blue'>Step 4: Final Equation</font>**

The equation of the regression line is:

$$
\hat{y} = 41.5 + 7.5x
$$

---

### **<font color='blue'>Interpretation</font>**

- **<font color='green'>Slope ($\theta_1 = 7.5$)</font>**: For every additional hour of study, the exam score increases by 7.5 points.  
- **<font color='green'>Intercept ($\theta_0 = 41.5$)</font>**: If a student studies for 0 hours, their predicted exam score is 41.5.

---

### **<font color='blue'>Prediction Example</font>**

If a student studies for 6 hours ($x = 6$), the predicted score is:

$$
\hat{y} = 41.5 + 7.5(6) = 41.5 + 45 = 86.5
$$

Thus, the predicted exam score is **86.5**.

---

## **<font color='darkorange'>Summary</font>**

1. The relationship between study hours and exam scores was modeled using **Simple Linear Regression**.  
2. The final regression equation is:

$$
\hat{y} = 41.5 + 7.5x
$$


# Implementation of SLR with formula

```python
# Input Data: Hours of Study (X) and Exam Scores (Y)
X = [1, 2, 3, 4, 5]  # Independent Variable
Y = [50, 55, 65, 70, 80]  # Dependent Variable

# Step 1: Calculate Mean of X and Y
n = len(X)
mean_X = sum(X) / n
mean_Y = sum(Y) / n

# Step 2: Calculate Slope (b) and Intercept (a)
numerator = sum((X[i] - mean_X) * (Y[i] - mean_Y) for i in range(n))
denominator = sum((X[i] - mean_X) ** 2 for i in range(n))

b = numerator / denominator  # Slope
a = mean_Y - b * mean_X      # Intercept

# Step 3: Display the Results
print(f"Mean of X: {mean_X}")
print(f"Mean of Y: {mean_Y}")
print(f"Slope (b): {b}")
print(f"Intercept (a): {a}")

# Final Regression Line Equation
print(f"Regression Line: Y = {a:.2f} + {b:.2f}X")

# Step 4: Prediction
def predict(x):
    return a + b * x

# Predict Y for a new value of X
X_new = 6
Y_pred = predict(X_new)
print(f"Predicted Y for X = {X_new}: {Y_pred:.2f}")
```

## Plotting

```python
import matplotlib.pyplot as plt
# Generate Predicted Y values for the regression line
Y_regression = [predict(x) for x in X]

# Plotting the data points and regression line
plt.figure(figsize=(8, 5))
plt.scatter(X, Y, color="blue", label="Actual Data", marker="o")  # Data points
plt.plot(X, Y_regression, color="red", label="Regression Line")  # Regression line
plt.scatter(X, Y_regression, color="orange", label="Regression Points", marker="x", s=100)  # Regression points

# Styling the plot
plt.title("Simple Linear Regression: Study Hours vs Exam Scores")
plt.xlabel("Hours of Study")
plt.ylabel("Exam Score")
plt.legend()
plt.grid(True)
plt.show()
```

# Implementation of SLR with sklearn

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Input Data: Hours of Study (X) and Exam Scores (Y)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent Variable (2D array for sklearn)
Y = np.array([50, 55, 65, 70, 80])  # Dependent Variable

# Step 1: Initialize and Train the Model
model = LinearRegression()
model.fit(X, Y)

# Step 2: Extract Slope (b) and Intercept (a)
a = model.intercept_  # Intercept
b = model.coef_[0]    # Slope

# Step 3: Display Results
print(f"Slope (b): {b}")
print(f"Intercept (a): {a}")
print(f"Regression Line: Y = {a:.2f} + {b:.2f}X")

# Step 4: Make Predictions
X_new = np.array([[6]])  # New value for prediction
Y_pred = model.predict(X_new)

print(f"Predicted Y for X = 6: {Y_pred[0]:.2f}")
```

# **<font color='darkorange'>The Normal Equation</font>**

The **Normal Equation** is a mathematical method to compute the parameters $(\theta)$ of a linear regression model that minimizes the cost function. This approach provides a direct closed-form solution, unlike iterative methods like Gradient Descent.

---

## **<font color='blue'>Normal Equation Formula</font>**

The Normal Equation is expressed as:  
$$
\theta = (X^T X)^{-1} X^T y
$$

### **<font color='green'>Explanation of Terms</font>**
- **$\theta$**: The vector of parameters, including the intercept $(\theta_0)$ and coefficients $(\theta_1, \theta_2, \ldots, \theta_n)$.
- **$X$**: The matrix of input features, where each row represents a data point and the first column is all 1s for the intercept.
- **$y$**: The vector of target values $(y^{(1)}$ to $y^{(m)})$.
- **$X^T$**: The transpose of the matrix $X$.
- **$(X^T X)^{-1}$**: The inverse of the matrix $X^T X$.

---
## **<font color='blue'>Deriving the Normal Equation for Linear Regression </font>**

In linear regression, we model the relationship between the input features <font color="blue">$x$</font> and the target variable <font color="blue">$y$</font> using a linear equation:

$$
\hat{y} = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \cdots + \theta_n x_n
$$

Where:
- <font color="blue">$y$</font> is the predicted output,
- <font color="blue">$x_1, x_2, \dots, x_n$</font> are the input features,
- <font color="blue">$\theta_0, \theta_1, \dots, \theta_n$</font> are the parameters (weights) to be learned.

### Matrix Notation

To simplify the equations, we express the linear regression model in **matrix notation**. We can write the inputs and parameters as vectors and the model equation becomes:

$$
\hat{y} = X \theta
$$

Where:
- <font color="blue">$y$</font> is an <font color="green">$m \times 1$</font> vector of target values (where <font color="green">$m$</font> is the number of training instances),
- <font color="blue">$X$</font> is an <font color="green">$m \times (n + 1)$</font> matrix of input features (with <font color="blue">$1$</font> added as the first feature for the bias term <font color="blue">$ \theta_0$</font>),
- <font color="blue">$ \theta$</font> is an <font color="green">$(n + 1) \times 1$</font> vector of model parameters.

### Cost Function

To find the best parameters, we need to minimize the error between the predicted values and the actual values. This is done using the **Sum of Squared Errors (SSE)** cost function, which is defined as:

$$
SSE(\theta) = \sum_{i=1}^{m} (y^{(i)} - \hat{y}^{(i)})^2
$$

Where:
- <font color="blue">$\hat{y}^{(i)}$</font> is the predicted output for the <font color="blue">$i$</font>-th training example,
- <font color="blue">$y^{(i)}$</font> is the true target value for the <font color="blue">$i$</font>-th training example,
- <font color="blue">$m$</font> is the number of training examples.

In matrix form, the cost function is:

$$
SSE(\theta) = (y - X \theta)^T (y - X \theta)
$$

### Step 1: Expand the SSE Expression

We begin with the cost function:

$$
SSE(\theta) = (y - X \theta)^T (y - X \theta)
$$

Expanding this expression:

$$
SSE(\theta) = y^T y - y^T X \theta - \theta^T X^T y + \theta^T X^T X \theta
$$

Since <font color="blue">$y^T X \theta$</font> is a scalar, we can rewrite it as <font color="blue">$\theta^T X^T y$</font>, and thus the expression becomes:

$$
SSE(\theta) = y^T y - 2 \theta^T X^T y + \theta^T X^T X \theta
$$

### Step 2: Derivative of the SSE

Now, let's differentiate the cost function with respect to <font color="blue">$ \theta$</font>.

- The derivative of <font color="blue">$y^T y$</font> with respect to <font color="blue">$ \theta$</font> is zero, as it is a constant.
- The derivative of <font color="blue">$-2 \theta^T X^T y$</font> with respect to <font color="blue">$ \theta$</font> is <font color="blue">$-2 X^T y$</font>.
- The derivative of <font color="blue">$ \theta^T X^T X \theta$</font> with respect to <font color="blue">$ \theta$</font> is <font color="blue">$2 X^T X \theta$</font>.

Thus, the derivative of the SSE is:

$$
\frac{\partial SSE(\theta)}{\partial \theta} = -2 X^T y + 2 X^T X \theta
$$

### Step 3: Set the Derivative Equal to Zero

To minimize the SSE, we set the derivative equal to zero:

$$
-2 X^T y + 2 X^T X \theta = 0
$$

Simplifying this expression:

$$
X^T X \theta = X^T y
$$

### Step 4: Solve for <font color="blue">$ \theta$</font>

Finally, we solve for <font color="blue">$ \theta$</font> by multiplying both sides by <font color="blue">$(X^T X)^{-1}$</font> (assuming <font color="blue">$X^T X$</font> is invertible):

$$
\theta = (X^T X)^{-1} X^T y
$$

This is the **Normal Equation**, and it gives us the closed-form solution to find the optimal model parameters <font color="blue">$ \theta$</font>.

## Key Points
- The **Normal Equation** involves computing the inverse of the matrix <font color="blue">$X^T X$</font>. This can be computationally expensive for large datasets with many features.
- If <font color="blue">$X^T X$</font> is not invertible (i.e., the matrix is singular or non-invertible), we cannot use the Normal Equation. In such cases, we might need to regularize the model (e.g., using Ridge Regression) or use an iterative method like **Gradient Descent**.


---
## **<font color='blue'>Steps to Solve Using the Normal Equation</font>**

1. **<font color='green'>Prepare the Data</font>**:
   - Ensure $X$ is a 2D array (add a column of 1s for the intercept).
   - Ensure $y$ is a 1D array of target values.

2. **<font color='green'>Compute $\theta$</font>**:
   - Transpose the matrix $X$ to get $X^T$.
   - Multiply $X^T$ by $X$.
   - Compute the inverse of $X^T X$.
   - Multiply the result by $X^T y$ to get $\theta$.
--

# **<font color='darkorange'>Coding Example</font>**

```python
import numpy as np

# Input Data: Hours of Study (X) and Exam Scores (Y)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent Variable (2D array)
Y = np.array([50, 55, 65, 70, 80])  # Dependent Variable


# Step 1: Add a column of 1s to X for the intercept
X_b = np.c_[np.ones((X.shape[0], 1)), X]  # Add a column of 1s to X

print(X_b)
print(Y)

# Step 2: Compute theta using the Normal Equation
theta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ Y

# Display the results
print(f"Optimal parameters (theta): {theta}")
print(f"Intercept (theta_0): {theta[0]}")
print(f"Slope (theta_1): {theta[1]}")

# Step 3: Prediction using the model
def predict(x_new):
    X_new = np.c_[np.ones((len(x_new), 1)), x_new]  # Add intercept column
    return X_new @ theta

# Predict Y for a new value of X
X_new = np.array([[6]])  # Predict for 6 hours of study
Y_pred = predict(X_new)
print(f"Predicted Exam Score for X = 6: {Y_pred[0]:.2f}")
```

```python
import numpy as np

# Input Data: Size (X1), Bedrooms (X2), Age (X3), Price (Y)
X = np.array([[1, 1500, 3, 15],
              [1, 1800, 4, 20],
              [1, 2400, 3, 10],
              [1, 3000, 4, 8],
              [1, 3500, 5, 5]])

Y = np.array([400000, 460000, 560000, 600000, 720000])

# Step 1: Compute theta using the Normal Equation
theta = np.linalg.inv(X.T @ X) @ X.T @ Y

# Display the results
print(f"Optimal parameters (theta): {theta}")
print(f"Intercept (theta_0): {theta[0]}")
print(f"Size coefficient (theta_1): {theta[1]}")
print(f"Bedrooms coefficient (theta_2): {theta[2]}")
print(f"Age coefficient (theta_3): {theta[3]}")

# Step 2: Prediction using the model
def predict(X_new):
    X_new = np.c_[np.ones((X_new.shape[0], 1)), X_new]  # Add intercept column
    return X_new @ theta

# Predict price for a new house with size 2500 sq ft, 3 bedrooms, and 10 years old
X_new = np.array([[2500, 3, 10]])
Y_pred = predict(X_new)
print(f"Predicted Price for X = {X_new[0]}: {Y_pred[0]:.2f}")
```

# <font color='darkorange'>Computational Complexity</font>

The **<font color='darkorange'>Normal Equation</font>** involves computing the inverse of $X^T X$, which is an $(n + 1) \times (n + 1)$ matrix, where $n$ is the number of features. The **<font color='red'>computational complexity</font>** of inverting such a matrix is typically about $O(n^{2.4})$ to $O(n^3)$, depending on the implementation.

In simpler terms, if you double the number of features, the computation time increases by approximately:

$$
2^{2.4} \approx 5.3 \quad \text{to} \quad 2^3 = 8
$$

Therefore, as the number of features grows, the **<font color='darkorange'>Normal Equation</font>** becomes computationally expensive. When dealing with datasets having a large number of features (e.g., 100,000), this approach may not be feasible.

## Gradient Descent: A Better Approach for Large Datasets

When the number of features is very large, or the dataset is too big to fit into memory, the **<font color='darkorange'>Normal Equation</font>** method become inefficient. In such cases, **<font color='darkorange'>Gradient Descent</font>** is a better option.
