```python
# Task 26-29: Example usage of external arrays
foo = [True, False, True, True, False, False, True, False, False, True]

print("\nRows where foo is True:\n", df[pd.Series(foo)]) # desplays the rows where true
```

```python
# 27. Subset rows where an external array, bar, is positive
bar = [1, -2, 3, -4, 5, 0, 2, -1, -2,4]
df[pd.Series(bar)>0]
```

```python
# 28. Subset rows where foo is TRUE or bar is negative
df[(pd.Series(foo)) | (pd.Series(bar)<0)]
```

```python
# 29. Subset the rows where foo is not TRUE and bar is not negative
df[(~pd.Series(foo)) & (pd.Series(bar)>=0)]
```
